Importing Numpy as np, Pandas as pd, Matplotlib as plt

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing CSV Files

```python
book_df = pd.read_csv('books.csv')
customers_df = pd.read_csv('customers.csv')
orders_df = pd.read_csv('orders.csv')
```

For Numpy Arrays

```python
books = book_df.values
customers = customers_df.values
orders = orders_df.values
```

Numpy Problems


Use NumPy to group by Book_ID and find the book with the maximum total sales (Quantity).

```python
Book_ID = orders[:,2].astype(int)
Quantities = orders[:,4].astype(int)
book_titles = books[:, 1]   # Title column
book_ids = books[:, 0].astype(int)
Author = books[:,2].astype(str)
unique_books, total_sales = np.unique(Book_ID, return_counts=False), np.zeros(len(np.unique(Book_ID)))
for i, book in enumerate(unique_books):
    total_sales[i] = Quantities[Book_ID == book].sum()
max_index = np.argmax(total_sales)
best_book_id = unique_books[max_index]
best_book_sales = total_sales[max_index]
print(f'Best selling book ID: {best_book_id}, Total quantity sold: {int(best_book_sales)}')

book_title = book_titles[book_ids == best_book_id][0]
print(f"Best selling book: '{book_title}' (ID: {best_book_id}), Quantity Sold: {int(best_book_sales)}")

Author = Author[book_ids == best_book_id][0]
print(f'Author of the book is: ', Author)
```


Convert Order_Date into months and use NumPy to sum Total_Amount per month.

```python
Order_Date = orders[:,3]
Total_Amount = orders[:,5].astype(float)
Order_Date = orders[:,3].astype(str)
Order_Month = np.array([d[:7]for d in Order_Date])


# Sum of Total_Amount per Month

unique_months = np.unique(Order_Month)
monthly_revenue = np.zeros(len(unique_months))
for i, month in enumerate(unique_months):
    monthly_revenue[i]= Total_Amount[Order_Month == month].sum()

for m, r in zip(unique_months, monthly_revenue):
    print(f'Month: {m}, Total Revenue: {r:.2f}')

```

Find how many unique customers (Customer_ID) have placed orders.

```python
customer_id = customers[:,0].astype(int)


# Find unique customers
unique_customers = np.unique(customer_id)

# Count of unique customers
num_unique_customers = unique_customers.size

print("Unique customers:", unique_customers)
print("Number of unique customers:", num_unique_customers)

```

Pandas Problem


Create a DataFrame and group by Customer_ID, calculate total amount spent, and find top 5 customers.


```python
merge_df = pd.merge(orders_df, customers_df, on='Customer_ID', how='inner')

final_merge_df = pd.merge(merge_df, book_df, on='Book_ID', how='inner')

Total_Amount_Spent = final_merge_df.groupby('Customer_ID')['Total_Amount'].sum().sort_values(ascending=False).reset_index()
Total_Amount_Spent.to_excel('Total_Amount_Spent.xlsx', index=False)
print(Total_Amount_Spent.head(5))
```

Merge orders and books, then group by Genre to find which genre has the highest sales.


```python
Highest_sales_by_genre = final_merge_df.groupby('Genre')['Total_Amount'].sum().sort_values(ascending=False).reset_index()
Highest_sales_by_genre.to_excel('Highest_sales_by_genre.xlsx', index=False)
print(Highest_sales_by_genre.head(1))
```

Merge customers and orders to compute total revenue per country.


```python
Total_revenue_Coutnry_wise = final_merge_df.groupby('Country')['Total_Amount'].sum().sort_values(ascending=False).reset_index()
Total_revenue_Coutnry_wise.to_excel('Total_revenue_Coutnry_wise.xlsx', index=False)
print(Total_revenue_Coutnry_wise.head(5))
```

Identify books in books that donâ€™t appear in orders.


```python
unsold_books = book_df[~book_df['Book_ID'].isin(orders_df['Book_ID'])]
print(unsold_books)
```

Group books by Published_Year and count how many were published each year.

```python
books_published_each_year = final_merge_df.groupby('Published_Year')['Title'].count().sort_values(ascending=False)

print(books_published_each_year)
```

Matplotlib Solutions

Line plot of total revenue (Total_Amount) per month.


```python
orders_df['Order_Date'] = pd.to_datetime(orders_df['Order_Date'])
orders_df['Order_Month'] = orders_df['Order_Date'].dt.to_period('M')

total_revenue_month_wise = orders_df.groupby('Order_Month')['Total_Amount'].sum().sort_values(ascending=False)
print(total_revenue_month_wise.head())

plt.Figure(figsize=(10,6))
total_revenue_month_wise.sort_index(inplace=True)
total_revenue_month_wise.plot(kind='line', figsize=(10,6), color='skyblue', marker='o')
plt.title('Total_Revenue_Month_Wise')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout
plt.grid(True)
plt.savefig('Total_Revenue_Month_wise.png', dpi=300)
plt.show()
```

Bar chart of top 10 books by quantity sold.


```python
Top_10Books_Sold = final_merge_df.groupby('Title')['Quantity'].sum().sort_values(ascending=False).head(10)
print(Top_10Books_Sold)

plt.Figure(figsize=(10,6))
Top_10Books_Sold.plot(kind='bar', color='lightblue')
plt.title('Top_10Books_Sold')
plt.xlabel('Books')
plt.ylabel('Quantity')
plt.savefig('Top_10Books_Sold.png')
plt.show()
```

Pie chart of total sales per genre.


```python
# Group by Genre and sum total sales
Total_Sales_Per_Genre = final_merge_df.groupby('Genre')['Total_Amount'].sum().sort_values(ascending=False)

# Plot pie chart
plt.figure(figsize=(10,6))
Total_Sales_Per_Genre.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Total Sales Per Genre')
plt.ylabel('')  # Remove default ylabel
plt.tight_layout()
plt.savefig('Total_Sales_Per_Genre.png')
plt.show()
```

Horizontal bar chart showing revenue contribution of each country.


```python
# Group by Country and sum total revenue
Revenue_Contribution_Country = (
    final_merge_df.groupby('Country')['Total_Amount']
    .sum()
    .sort_values(ascending=True)  # ascending=True so largest at top in horizontal bar
)

# Plot horizontal bar chart
plt.figure(figsize=(10,60))
Revenue_Contribution_Country.plot(kind='barh', color='skyblue')
plt.title('Revenue Contribution by Country')
plt.xlabel('Total Revenue')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('Revenue_Contribution_Country.png')
plt.show()
```

Histogram of Published_Year to see book publishing distribution.

```python
# Ensure Published_Year is numeric
final_merge_df['Published_Year'] = pd.to_numeric(final_merge_df['Published_Year'], errors='coerce')

plt.figure(figsize=(10,6))
plt.hist(final_merge_df['Published_Year'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Book Publication Years')
plt.xlabel('Published Year')
plt.ylabel('Number of Books')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.savefig('Distribution of Book Publication Years.png')
plt.show()
```
